1. Merge flights with airlines on the column carrier, and save the result to flights_airlines.

4. Merge planes and flights tables with a right join on tailnum and add suffixes, where '_pln' and 'None (look up the documentation!) are suffixes for the left and right tables, respectively. Save the result as planes_flights.
From planes_flights, subset only the rows where the model column (or any other column from planes table) is not null. Keep all columns. Save the result as flights_model.
Use value_counts() on the column manufacturer of flights_model to find the distribution fo the plan types. Save the result as mfg_distribution
Group flights_model by model and count the number of tailnum values. Save the result as model_count, which should be a DataFrame object and its index column is model.
Sort the model_count by the count in descending order and save the result as model_sorted. Hint: 'model' is both an index level and a column label, which is ambiguous. So, renmae the column to m_count before sorting.

```python
planes_flights = planes.merge(
    flights,
    on='tailnum',
    how='right',
    suffixes=('_pln', None))
flights_model = planes_flights[planes_flights['model'].notna()]
mfg_distribution = flights_model['manufacturer'].value_counts()
model_count = flights_model.groupby('model')['tailnum'].count().to_frame()
model_count = model_count.rename(columns={'tailnum': 'm_count'})
model_sorted = model_count.sort_values(by='m_count', ascending=False)


display(planes_flights)
planes.info()
display(mfg_distribution)
display(model_count.iloc[[0,1,2,-3,-2,-1],:])
model_sorted.head()
```
