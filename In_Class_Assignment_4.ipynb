```python
birth_college_statements = []
for idx in nfl.index:
    name = nfl.at[idx, 'Name']
    birth_place = nfl.at[idx, 'Birth Place']
    college = nfl.at[idx, 'College']
    if pd.notna(birth_place):
        birth_college_statements.append(f"Player {name} from {birth_place} went to {college}.")
    else:
        birth_college_statements.append(f"Player {name} from unknown birth place went to {college}.")
print(birth_college_statements[5])
print(birth_college_statements[6])
```

6. Use a for loop to calculate BMI for each player and append it to bmi_list created below. Use the formula at the CDC website to use the US customary units (pounds and inches).
Add a new column named "BMI" with bmi_list.
Store the first 10 values of the "BMI" column in a list named first_10_bmi.

```python

bmi_list = []
for idx, row in nfl.iterrows():
    height = row['Height (inches)']
    weight = row['Weight (lbs)']
    if pd.notna(height) and pd.notna(weight) and height != 0:
        bmi = (weight / (height ** 2)) * 703
    else:
        bmi = np.nan
    bmi_list.append(bmi)
nfl['BMI'] = bmi_list
first_10_bmi = nfl['BMI'].head(10).tolist()
print(nfl.columns)
print(nfl['BMI'].head(10))
```

read the whole content of f and store it in a variable named content, and
